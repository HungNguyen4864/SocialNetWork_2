```python
import pandas as pd
from ast import literal_eval

# Load the Excel file
file_path = 'all_post_modified.xlsx'
data = pd.read_excel(file_path)

# Replace NaN values with empty lists in the columns that will be parsed
data['cleaned_reactions'] = data['cleaned_reactions'].apply(lambda x: literal_eval(x) if pd.notna(x) else [])
data['list_comments'] = data['list_comments'].apply(lambda x: literal_eval(x) if pd.notna(x) else [])

# Define a function to extract user IDs from lists of tuples (user_id, user_name)
def extract_user_ids(data_series):
    return [user_id for user_ids in data_series for user_id, _ in user_ids]

# Extract user IDs from cleaned_reactions and list_comments separately
data['reaction_user_ids'] = data['cleaned_reactions'].apply(lambda x: [user_id for user_id, _ in x])
data['comment_user_ids'] = data['list_comments'].apply(lambda x: [user_id for user_id, _ in x])

# Create a DataFrame with separate columns for post_user, reaction_user_ids, and comment_user_ids
separate_user_ids_df = data[['post_user', 'reaction_user_ids', 'comment_user_ids']]

# Display the first few rows of the new DataFrame
separate_user_ids_df.to_excel
```

```python
import ast
import networkx as nx

# Initialize a directed graph
G = nx.DiGraph()
file_path = 'separate_user_ids.xlsx'
data = pd.read_excel(file_path)

# Function to safely parse lists from string representation
def parse_list(data_str):
    try:
        return ast.literal_eval(data_str)
    except:
        return []

# Process each row to build the graph
for index, row in data.iterrows():
    post_user = row['post_user']
    reaction_user_ids = parse_list(row['reaction_user_ids'])
    comment_user_ids = parse_list(row['comment_user_ids'])
    
    # Add edges from post_user to reaction and comment users
    for user_id in reaction_user_ids:
        G.add_edge(post_user, user_id)
    for user_id in comment_user_ids:
        G.add_edge(post_user, user_id)
    
    # Add edges between reaction and comment users
    for reaction_user in reaction_user_ids:
        for comment_user in comment_user_ids:
            G.add_edge(reaction_user, comment_user)
            G.add_edge(comment_user, reaction_user)

# Output basic graph info and number of nodes and edges
graph_info = print(G)
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

graph_info, num_nodes, num_edges
```

```python
edges_df = pd.DataFrame(list(G.edges()), columns=['Source', 'Target'])

# Save the DataFrame to an Excel file
output_file_path = 'graph_edges.csv'
edges_df.to_csv(output_file_path, index=False)

output_file_path
```
