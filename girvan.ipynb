```python
import networkx as nx
from networkx.algorithms.community import girvan_newman
import matplotlib.pyplot as plt
```

```python
G = nx.les_miserables_graph()
print(f"So dinh: {G.number_of_nodes()}")
print(f"So canh: {G.number_of_edges()}")
communities = girvan_newman(G)

communities = list(communities)
print(f"So la lap toi da: {len(communities)}")
```

```python
def draw_graph(G, show_names=False, node_size=50):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=show_names, node_size=node_size, font_size=8)
    plt.show()
```

```python
selected_communities = communities[9]
```

```python
for i, community in enumerate(selected_communities):
    G_community = G.subgraph(community)
    # draw_graph(G_community)
    print(f"community {i}: {community}")
    draw_graph(G_community,show_names=True,node_size=100)
```
